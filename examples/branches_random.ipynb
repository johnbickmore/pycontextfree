```python
import math
from contextfree.contextfree import *
```

```python
@check_limits
def branch():
    circle(1)
    with translate(0,0.6):
        with color(alpha=0.997):
            with rotate(0.04):
                with scale(0.998):
                    if coinflip(50):
                        with flip_y():
                            branch()
                    else:
                        branch()
            
            if coinflip(60):
                with flip_y():
                    with scale(0.9):
                        branch()
    
init(canvas_size=(400, 400), face_color="#44cc00", background_color="#082019", max_depth=800)
branch()
display_ipython()
```
