```python
import math
from contextfree import *
```

```python
@check_limits
def branch():
    circle(1)
    with transform(y=0.6, alpha=-0.2, angle=0.3):
        with scale(0.998):
            if coinflip(50):
                 with flip_y():
                    branch()
            else:
                 branch()
            
            if coinflip(120):
                with flip_y():
                    with scale(0.9):
                        branch()
    
init(canvas_size=(400, 400), face_color="#44cc00", background_color="#082019", max_depth=700)
branch()
display_ipython()
```
