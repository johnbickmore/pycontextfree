```python
def sheet():
    with color(lightness=0.1, alpha=0.03):
        with scale(1.3):
            box()
    with color(lightness=0.1):
        with scale(1.007):
            box()    
    box()

@check_limits
def pyramid():
    sheet()
    with rotate(0.09 + rnd(0.02)):
        with scale(0.99):
            with translate(rnd(0.07), rnd(0.07)):
                with color(hue=0.005+rnd(0.001)):
                    pyramid()
        #with translate(0,0.9):
     #   with scale(0.7 + rnd(0.15)):
      #      with rotate(-0.3 + rnd(0.3)):
       #         branch()
        #    with rotate(0.3 + rnd(0.3)):
         #       branch()


init(canvas_size=(400,400), face_color="#784312", max_depth=4000)
with scale(0.6):
    pyramid()
        
display_ipython()
```
