```python
from contextfree import *
```

```python
def sheet():
    with transform(lightness=-1, alpha=-0.97, scale_x=1.3):
        box()
    with color(lightness=-1):
        with scale(1.007):
            box()    
    box()

@check_limits
def pyramid():
    sheet()
    with transform(x=rnd(0.08), y=rnd(0.08), scale_x=0.99, angle=prnd(1.2), hue=rnd(3)):
        pyramid()

init(canvas_size=(400,400), face_color="#48d341", max_depth=10000)
with transform(scale_x=2, angle=prnd(90), hue=rnd(100)):
    pyramid()
        
display_ipython()
```
